Feature engineering

# --------------------------------------------
# Encoding Important Features (Mapping In-Place)
# --------------------------------------------

```python
# Define mapping dictionaries
age_group_mapping = {'Adult': 1, 'Child': 0}
sex_mapping = {'Male': 0, 'Female': 1}
living_situation_mapping = {
    'PRIVATE RESIDENCE': 1,
    'INSTITUTIONAL SETTING': 2,
    'OTHER LIVING SITUATION': 0,
    'UNKNOWN': -1
}
household_composition_mapping = {
    'COHABITATES WITH OTHERS': 1,
    'LIVES ALONE': 0,
    'NOT APPLICABLE': -1
}

employment_status_mapping = {
    'EMPLOYED': 2,
    'UNEMPLOYED, LOOKING FOR WORK': 1,
    'NOT IN LABOR FORCE: UNEMPLOYED AND NOT LOOKING FOR WORK': 0
}
hours_worked_mapping = {
    'NOT APPLICABLE': 0,
    '01-14 HOURS': 1,
    '15-34 HOURS': 2,
    '35 HOURS OR MORE': 3,
    'UNKNOWN EMPLOYMENT HOURS': -1
}
education_status_mapping = {
    'NO FORMAL EDUCATION': 0,
    'PRE-K TO FIFTH GRADE': 1,
    'MIDDLE SCHOOL TO HIGH SCHOOL': 2,
    'SOME COLLEGE': 3,
    'COLLEGE OR GRADUATE DEGREE': 4,
    'OTHER': -1,
    'UNKNOWN': -1
}
race_mapping = {
    'White only': 1,
    'Black only': 2,
    'Multi-racial': 3,
    'Other': 0,
    'Unknown': -1
}
hispanic_mapping = {
    'YES, HISPANIC/LATINO': 1,
    'NO, NOT HISPANIC/LATINO': 0
}

binary_mapping = {
    'Yes': 1,
    'No': 0,
    'Unknown': -1,
    'UNKNOWN': -1
}

# Apply mappings
df['Age Group'] = df['Age Group'].map(age_group_mapping)
df['Sex'] = df['Sex'].map(sex_mapping)
df['Living Situation'] = df['Living Situation'].map(living_situation_mapping)
df['Household Composition'] = df['Household Composition'].map(household_composition_mapping)
df['Employment Status'] = df['Employment Status'].map(employment_status_mapping)
df['Number Of Hours Worked Each Week'] = df['Number Of Hours Worked Each Week'].map(hours_worked_mapping)
df['Education Status'] = df['Education Status'].map(education_status_mapping)
df['Race'] = df['Race'].map(race_mapping)
df['Hispanic Ethnicity'] = df['Hispanic Ethnicity'].map(hispanic_mapping)

# Apply binary mapping to all other Yes/No columns
binary_columns = [
    'Mental Illness', 'Intellectual Disability', 'Autism Spectrum', 'Other Developmental Disability',
    'Alcohol Related Disorder', 'Drug Substance Disorder', 'Opioid Related Disorder',
    'Mobility Impairment Disorder', 'Hearing Impairment', 'Visual Impairment',
    'Speech Impairment', 'Hyperlipidemia', 'High Blood Pressure', 'Diabetes',
    'Obesity', 'Heart Attack', 'Stroke', 'Other Cardiac', 'Pulmonary Asthma',
    'Alzheimer or Dementia', 'Kidney Disease', 'Liver Disease', 'Endocrine Condition',
    'Neurological Condition', 'Traumatic Brain Injury', 'Joint Disease', 'Cancer',
    'Other Chronic Med Condition', 'No Chronic Med Condition', 'Unknown Chronic Med Condition',
    'Cannabis Recreational Use', 'Cannabis Medicinal Use', 'Smokes',
    'Received Smoking Medication', 'Received Smoking Counseling', 'Serious Mental Illness',
    'Alcohol 12m Service', 'Opioid 12m Service', 'Drug/Substance 12m Service',
    'SSI Cash Assistance', 'SSDI Cash Assistance', 'Veterans Disability Benefits',
    'Veterans Cash Assistance', 'Public Assistance Cash Program', 'Other Cash Benefits',
    'Medicaid and Medicare Insurance', 'No Insurance', 'Unknown Insurance Coverage',
    'Medicaid Insurance', 'Medicaid Managed Insurance', 'Medicare Insurance',
    'Private Insurance', 'Child Health Plus Insurance', 'Other Insurance'
]

# Apply binary mapping
for col in binary_columns:
    df[col] = df[col].map(binary_mapping)

print("All mappings applied successfully!")
```

Feature Engineering: Dropping + Combining Features

```python
# 1. Drop Irrelevant Columns
columns_to_drop = [
    'Survey Year',
    'Region Served',
    'Transgender',
    'Sexual Orientation',
    'Preferred Language',
    'Religious Preference',
    'Principal Diagnosis Class',
    'Additional Diagnosis Class',
    'Three Digit Residence Zip Code'
]

df.drop(columns=columns_to_drop, axis=1, inplace=True)
print(f"Dropped {len(columns_to_drop)} irrelevant columns.")
```

```python
# 2. Create New Combined Features

# Addiction Score: Sum of substance use disorder indicators
df['Addiction_Score'] = (
    df['Alcohol Related Disorder'] +
    df['Drug Substance Disorder'] +
    df['Opioid Related Disorder']
)

# Disability Score: Sum of physical disability indicators
df['Disability_Score'] = (
    df['Mobility Impairment Disorder'] +
    df['Hearing Impairment'] +
    df['Visual Impairment'] +
    df['Speech Impairment']
)

# Chronic Condition Score: Sum of chronic medical conditions
df['Chronic_Condition_Score'] = (
    df['Hyperlipidemia'] +
    df['High Blood Pressure'] +
    df['Diabetes'] +
    df['Obesity'] +
    df['Heart Attack'] +
    df['Stroke'] +
    df['Other Cardiac'] +
    df['Pulmonary Asthma']
)

# Cannabis Usage Score: Recreational + Medicinal
df['Cannabis_Usage_Score'] = (
    df['Cannabis Recreational Use'] +
    df['Cannabis Medicinal Use']
)

# Cash Assistance Score: Sum of cash benefits
df['Cash_Assistance_Score'] = (
    df['SSI Cash Assistance'] +
    df['SSDI Cash Assistance'] +
    df['Veterans Disability Benefits'] +
    df['Veterans Cash Assistance'] +
    df['Public Assistance Cash Program'] +
    df['Other Cash Benefits']
)

# Insurance Coverage Score: Sum of different insurance types
df['Insurance_Coverage_Score'] = (
    df['Medicaid and Medicare Insurance'] +
    df['Medicaid Insurance'] +
    df['Medicaid Managed Insurance'] +
    df['Medicare Insurance'] +
    df['Private Insurance'] +
    df['Child Health Plus Insurance'] +
    df['Other Insurance']
)

print("Created new combined features successfully!")



```

```python
# 3. Drop the original columns used for combination if not needed individually anymore
columns_to_drop_after_combining = [
    'Alcohol Related Disorder', 'Drug Substance Disorder', 'Opioid Related Disorder',
    'Mobility Impairment Disorder', 'Hearing Impairment', 'Visual Impairment', 'Speech Impairment',
    'Hyperlipidemia', 'High Blood Pressure', 'Diabetes', 'Obesity', 'Heart Attack', 'Stroke', 'Other Cardiac', 'Pulmonary Asthma',
    'Cannabis Recreational Use', 'Cannabis Medicinal Use',
    'SSI Cash Assistance', 'SSDI Cash Assistance', 'Veterans Disability Benefits', 'Veterans Cash Assistance',
    'Public Assistance Cash Program', 'Other Cash Benefits',
    'Medicaid and Medicare Insurance', 'Medicaid Insurance', 'Medicaid Managed Insurance',
    'Medicare Insurance', 'Private Insurance', 'Child Health Plus Insurance', 'Other Insurance'
]

df.drop(columns=columns_to_drop_after_combining, axis=1, inplace=True)
print(f"Dropped {len(columns_to_drop_after_combining)} columns after combining.")
```

```python
# Step 1: Select important features (scaled_df)
selected_features_svm = [
    'Age Group', 'Sex', 'Race', 'Hispanic Ethnicity',
    'Household Composition', 'Living Situation', 'Employment Status',
    'Number Of Hours Worked Each Week', 'Education Status',
    'Mental Illness', 'Intellectual Disability', 'Autism Spectrum',
    'Other Developmental Disability', 'Alzheimer or Dementia', 'Kidney Disease',
    'Liver Disease', 'Endocrine Condition', 'Neurological Condition',
    'Traumatic Brain Injury', 'Joint Disease', 'Cancer',
    'Addiction_Score', 'Disability_Score', 'Chronic_Condition_Score',
    'Cannabis_Usage_Score', 'Cash_Assistance_Score', 'Insurance_Coverage_Score'
]

X_svm = scaled_df[selected_features_svm]
y_svm = scaled_df['Program Category Encoded']
```

```python


# Step 2: Train-Test Split
X_train_svm, X_test_svm, y_train_svm, y_test_svm = train_test_split(X_svm, y_svm, test_size=0.2, random_state=42)
```

```python
X_train_svm = X_train_svm.fillna(0)
X_test_svm = X_test_svm.fillna(0)
```

```python
# Combine X and y temporarily
train_data = pd.concat([X_train_svm, y_train_svm], axis=1)
test_data = pd.concat([X_test_svm, y_test_svm], axis=1)

# Drop rows where y is NaN
train_data = train_data.dropna(subset=['Program Category Encoded'])
test_data = test_data.dropna(subset=['Program Category Encoded'])

# Split back X and y
X_train_svm = train_data.drop(columns=['Program Category Encoded'])
y_train_svm = train_data['Program Category Encoded']

X_test_svm = test_data.drop(columns=['Program Category Encoded'])
y_test_svm = test_data['Program Category Encoded']

print("Dropped rows with NaN target labels. Ready for SVM training.")
```

```python

#  Step 3: Train initial SVM
svm_model = SVC(random_state=42)
svm_model.fit(X_train_svm, y_train_svm)
```

```python

#  Step 4: Predictions
y_pred_svm = svm_model.predict(X_test_svm)
```

```python
#  Step 5: Evaluation
print("Initial SVM Classification Report:")
print(classification_report(y_test_svm, y_pred_svm))
```

```python
print("Confusion Matrix:")
print(confusion_matrix(y_test_svm, y_pred_svm))
```

```python
# Step 6: Hyperparameter Tuning
param_grid_svm = {
    'C': [0.1, 1, 10],
    'kernel': ['linear', 'rbf', 'poly'],
    'gamma': ['scale', 'auto']
}
grid_svm = GridSearchCV(SVC(random_state=42), param_grid_svm, cv=5, scoring='accuracy', verbose=2, n_jobs=-1)
grid_svm.fit(X_train_svm, y_train_svm)

print(f"Best Parameters (SVM): {grid_svm.best_params_}")
```

```python
# Retrain using best parameters
best_svm_model = grid_svm.best_estimator_
```

```python
# Step 7: Final Evaluation
y_pred_best_svm = best_svm_model.predict(X_test_svm)
```

```python
print("Final SVM Classification Report (after tuning):")
print(classification_report(y_test_svm, y_pred_best_svm))
```

```python

print("Final Confusion Matrix:")
print(confusion_matrix(y_test_svm, y_pred_best_svm))
```
